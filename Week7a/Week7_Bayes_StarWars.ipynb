```python
from scipy.stats import beta
import matplotlib.pyplot as plt
import numpy as np
```

Recall that the Beta distribution consists of an alpha and beta - where the alpha represents the number of observed successes and beta represents the number of observed failures

```python
from scipy.stats.stats import skew
a, b = 2, 7440

mean, var, skew, kurt = beta.stats(a, b, moments='mvsk')
```

Let's imagine that C3PO's assumption is based off his knowledge that 2 people actually made an astroid run while 7,440 people did not
StarWars2.PNG

```python
fig, ax = plt.subplots(1,1)

x = np.linspace(beta.ppf(0.01, a, b),
                beta.ppf(0.99, a, b), 100)
ax.plot(x, beta.pdf(x, a, b),
       'r-', lw=5, alpha=0.6, label='beta pdf')

fig.suptitle('C3P0 Likelihood of Surviving the Astroid Run')
plt.xlabel('Probability of Success')
plt.ylabel('Density')
```

Those are some pretty low odds - but we know that Han is a pretty good pilot - so we can model our **prior** probability that Han and his "mad" skills (sorry for the mixed pop culture references) will make it through the astroid field.  Again using our Beta distribution, we think Han will be successful 20,000 times to 1

```python
ah, bh = 20000, 1
fig, ax = plt.subplots(1,1)

x = np.linspace(beta.ppf(0.01, ah, bh),
                beta.ppf(0.99, ah, bh), 100)
ax.plot(x, beta.pdf(x, ah, bh),
       'r-', lw=5, alpha=0.6, label='beta pdf')

fig.suptitle('Prior Belief of Han Making the Astroid Run')
plt.xlabel('Probability of Success')
plt.ylabel('Density')
```

Bayes Theorem tells us that our Posterior propbability is the product of our likelihood (C3PO's data) and our prior (belief in Han) - thankfully the simplicity of the Beta distribution allows us to add the alpha of our likelihood (a) and the alpha of our prior (ah) to get our alpha posterior (ab below).  And a similar calculation for our beta.

```python
ab, bb = 20002, 7401
fig, ax = plt.subplots(1,1)

x = np.linspace(beta.ppf(0.01, ab, bb),
                beta.ppf(0.99, ab, bb), 100)
ax.plot(x, beta.pdf(x, ab, bb),
       'r-', lw=5, alpha=0.6, label='beta pdf')

fig.suptitle('Bayesian Posterior of the Astroid Run')
plt.xlabel('Probability of Success')
plt.ylabel('Density')
```
